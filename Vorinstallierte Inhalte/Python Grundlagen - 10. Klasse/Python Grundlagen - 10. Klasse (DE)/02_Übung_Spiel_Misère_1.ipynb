## 1.2: Schleifen: for und while <a class="anchor" id="chap_2"></a>

Wenn du Code schreibst, dann wirst du feststellen, dass manche Stücke des Codes häufig wiederholt werden müssten. Damit du dich in solchen Situationen nicht immer wieder wiederholen musst, gibt es  Schleifen. Die beiden häufigsten Typen von Schleifen in Python sind ***for*** und ***while***. Eine for-Schleife führt die Befehle für jeden Eintrag in einer Liste aus, und die while-Schleife läuft so lange, bis eine bestimmte Bedingung nicht mehr erfüllt ist. Stell dir den Code wie einen Schaffner im Zug vor: bei einer for-Schleife kontrolliert der Schaffner jeden Passagier, bis alle Passagiere kontrolliert sind. Bei einer while-Schleife kontrolliert er Passagiere so lange, bis sein Arbeitstag zu Ende ist - unabhängig davon, wie viele Passagiere er letztendlich kontrolliert hat. 

#### for-Schleifen

For-Schleifen haben eine Laufvariable, die als Wert nacheinander jeden Eintrag in einer Liste annimmt. Der Schleifenbefehl sieht so aus:

    for laufvariable in list:
        # Befehle
        
also zum Beispiel 

    for a in [1,2,3,4,5]:
        print(a)

Laufvariablen in Schleifen haben oft kurze Bezeichner (***a***) und es ist guter Programmierstil, Variablennamen wie ***a*** oder ***b*** nur für Laufvariablen zu verwenden, damit später keine Verwirrung entsteht. Die Laufvariable ändert sich nacheinander zu jedem Feld der Liste.

#### while-Schleifen

While-Schleifen wiederholen den Code in der Schleife so lange, bis die Bedingung im Schleifenkopf nicht mehr erfüllt ist. Damit können wir z.B. sicherstellen, dass der Spieler bei Auswahl des Münzstapels eine gültige Wahl trifft: "So lange die Eingabe nicht 1,2,3 oder 4 ist, wiederhole die Eingabe".

**🚀 &nbsp;Aufgabe 7**: Probiere den Code aus. Das "not in" bedeutet hier "nicht in", also "So lange Wert nicht in Liste".

```python
welcher_stapel = 0 # wir müssen die variable hier definieren, da sie im Schleifenkopf verwendet wird.
while welcher_stapel not in [1,2,3,4]:
    welcher_stapel = int(input("Von welchem Stapel ziehen?"))
print("Stapel", welcher_stapel, "ausgewählt.")
```

## 1.4: Funktionen definieren: Unsere Codestücke bekommen Namen <a class="anchor" id="chap_4"></a>

Funktionen erlauben es, ein Stück Code wie einen Befehl aufzurufen und somit an verschiedenen Stellen im Programm zu verwenden. Funktionen werden in Python mit dem Befehl ***def*** definiert. Und sie können optional mit dem Befehl ***return*** einen Wert oder mehrere Werte zurückliefern, der dann als Ergebnis verwendet werden kann.

## 1.5: Funktionen und lokale Variablen  <a class="anchor" id="chap_5"></a>

Innerhalb einer Funktion in Python sind die Variablen des Codes, der außerhalb der Funktion steht, erreichbar. Im Gegenzug sind die Variablen, die innerhalb einer Funktion erstellt werden, nicht von außerhalb erreichbar. 

**🚀 &nbsp;Aufgabe 13**: Führe das Codestück aus. Warum erzeugt es eine Fehlermeldung?

```python
a = 5
def gib_a_zurück():
    b = 3
    return a

gib_a_zurück()
print(b)
```

    <summary style="border: 1px solid; border-radius: 3px; padding: 5px; display: inline-block; cursor: pointer; margin-top: 10px; background-color: #EEE9E9;"> 🔎 Erklärung anzeigen</summary>
    
    Während die Variable a in der Funktion erreichbar ist und keinen Fehler erzeugt, 
    weiß das Hauptprogramm nichts von der Varible b, da diese nur innerhalb der Funktion definiert ist. 
    
</details>

<div class="alert alert-block alert-success">
<p style="color:black;font-size:1.5em;font-weight:bold">💡 Gut zu wissen: Globale Variablen sollten vermieden werden</p>
<font color="black" size="3">
<ul>
<li>Es ist guter Programmierstil, so wenig wie möglich globale Variablen einzusetzen, also Variablen, die im Hauptprogramm definiert werden. Grund ist, dass bei komplexen Programmen schlecht nachvollziehbar ist, in welcher Funktion eine Variable überschrieben wurde, und das Finden von Fehlern somit erschwert wird.</li>
<br>        
<li>Im Idealfall "lebt" ein Großteil der Variablen in Funktionen, und Funktionen verändern nur Variablen, die auch im Inneren der Funktion definiert sind.</li>
<br>
<li>Die "Kommunikation" zwischen den Funktionen erfolgt dann via Parameter und Rückgabewerte.</li>
</ul>
</font>
